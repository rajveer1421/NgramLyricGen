### Tokenizing And Preprocessing of the text

```python
def preprocess_string(s):
    # Remove all non-word characters (everything except numbers and letters)
    s = re.sub(r"[^\w\s]", '', s)
    # Replace all runs of whitespaces with no space
    s = re.sub(r"\s+", '', s)
    # replace digits with no space
    s = re.sub(r"\d", '', s)

    return s
```

```python
def process(words):
    tokens=word_tokenize(words)
    processed=[preprocess_string(token) for token in tokens]
    processed=[process.lower() for process in processed]
    return processed
```

```python
tokens=process(song)
print(tokens)
```

### Knowing the frequency of each word 

```python
fdist=nltk.FreqDist(tokens)
fdist
```

```python
plt.bar(list(fdist.keys())[0:10],list(fdist.values())[0:10])
plt.xlabel("Words")
plt.ylabel("Frequencies")
```

### Using Unigram Model

```python
total=sum(fdist.values())
total
```

### Predicting Frequency of word stranger

```python
fdist["strangers"]
```

```python
prob_stranger=fdist["strangers"]/total
```

```python
print("Probability of stranger in the list is", prob_stranger)
```

## Working With Bigram Model

```python
bigrams=nltk.bigrams(tokens)
```

```python
bigrams=list(bigrams)
```

```python
bigrams[0:10]
```

### Finding probability of  a given word given the previous word

```python
word1=input("Enter the word whose probability needs to be find")

```

```python
fbigram=nltk.FreqDist(bigrams)
```

```python
vocab_prob={}
for token in tokens:
    vocab_prob[token]=fbigram[token,word1]/fdist[word1]
vocab_prob = sorted(vocab_prob.items(), key=lambda x: x[1], reverse=True)
```
